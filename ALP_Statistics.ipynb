### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data_ex3 = {
    'x1': [1.74, 6.32, 6.22, 10.52, 1.19, 1.22, 4.10, 6.32, 4.08, 4.15, 10.15, 1.72, 1.70],
    'x2': [5.30, 5.42, 8.41, 4.63, 11.60, 5.85, 6.62, 8.72, 4.42, 7.60, 4.83, 3.12, 5.30],
    'x3': [10.8, 9.4, 7.2, 8.5, 9.4, 9.9, 8.0, 9.1, 8.7, 9.2, 9.4, 7.6, 8.2],
    'y':  [25.5, 31.2, 25.9, 38.4, 18.4, 26.7, 26.4, 25.9, 32.0, 25.2, 39.7, 35.7, 26.5]
}

df = pd.DataFrame(data_ex3)
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values  
result = LinearRegression()
result.fit(X, y)


print("--- Part (a): Model Estimation ---")
print(f"Intercept: {result.intercept_:.4f}")
print(f"Coefficient b1 (for x1): {result.coef_[0]:.4f}")
print(f"Coefficient b2 (for x2): {result.coef_[1]:.4f}")
print(f"Coefficient b3 (for x3): {result.coef_[2]:.4f}")
print(f"\nRegression Equation: y = {result.intercept_:.4f} + ({result.coef_[0]:.4f})*x1 + ({result.coef_[1]:.4f})*x2 + ({result.coef_[2]:.4f})*x3")


print("\n--- Part (b): Root Mean Squared Error (RMSE) ---")
y_pred = result.predict(X)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(f"Mean Squared Error (MSE): {mse:.4f}")
print('Root Mean Squared Error (RMSE): {:0.4f}'.format(np.sqrt(mean_squared_error(y, y_pred))))
```

```python
X = df_housing.iloc[:, :-1].values
# Selects all rows and only the last column from df_housing as the target variable (y).
# .values converts the DataFrame slice to a NumPy array.
# Last column is the one we want to predict

y = df_housing.iloc[:, -1].values
# Selects all rows and all columns except the last one from df_housing as features (X).
# .values converts the DataFrame slice to a NumPy array.
```

```python
result = LinearRegression()
# Initializes a Linear Regression model object.
result.fit(X, y)
# Fits (trains) the Linear Regression model using the features X and the target variable y.

print("Intercept: {:0.4f}".format(result.intercept_))
# Explanation of Output: This is the constant term (b0) in the regression equation.
# It's the predicted house price if all independent variables were zero.
# In this context, a large negative intercept might not be directly interpretable if X=0 is unrealistic.

print("Coefficient b1 (X1 transaction date): {:0.4f}".format(result.coef_[0]))
# Explanation of Output: For each one-unit increase in 'transaction date',
# the house price is predicted to increase by 5.1462 units, assuming other factors remain constant.

print("Coefficient b2 (X2 house age): {:0.4f}".format(result.coef_[1]))
# Explanation of Output: For each one-unit increase in 'house age',
# the house price is predicted to decrease by 0.2697 units, assuming other factors remain constant.


print("Coefficient b3 (X3 distance to the nearest MRT station): {:0.4f}".format(result.coef_[2]))
# Explanation of Output: For each one-unit increase in 'distance to the nearest MRT station',
# the house price is predicted to decrease by 0.0045 units, assuming other factors remain constant.

print("Coefficient b4 (X4 number of convenience stores): {:0.4f}".format(result.coef_[3]))
# Output: Coefficient b4 (X4 number of convenience stores): 1.1333
# Explanation of Output: For each additional 'convenience store',
# the house price is predicted to increase by 1.1333 units, assuming other factors remain constant.

print("Coefficient b5 (X5 latitude): {:0.4f}".format(result.coef_[4]))
# Explanation of Output: For each one-unit increase in 'longitude',
# the house price is predicted to decrease by 12.4236 units, assuming other factors remain constant.

print("Coefficient b6 (X6 longitude): {:0.4f}".format(result.coef_[5]))
# Explanation of Output: For each one-unit increase in 'longitude',
# the house price is predicted to decrease by 12.4236 units, assuming other factors remain constant.

print('\n Regression Equation: y = {:0.4f}*X1_transaction_date + {:0.4f}*X2_house_age + {:0.4f}*X3_distance_MRT + {:0.4f}*X4_convenience_stores + {:0.4f}*X5_latitude + {:0.4f}*X6_longitude + {:0.4f}'
      .format(result.coef_[0], result.coef_[1], result.coef_[2], result.coef_[3], result.coef_[4], result.coef_[5],result.intercept_))


# --- Display the Coefficient of Determination ---
print('\nCoefficient of Determination (R-squared) is {:0.4f}'.format(result.score(X, y)))
# Explanation of Output: This means that approximately 58.24% of the variability
# in house prices (y) can be explained by the linear model with these six features.
# The remaining 41.76% is due to other factors not included in the model or random variation.
```
